# Data visualisation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#on représente différentes courbes sur un graphique
plt.plot(np.random.randn(10),"b:",label="courbe1")
plt.plot(np.random.randn(10),"r--",label="courbe2")
plt.plot(np.random.randn(10),"y-",label="courbe3")
plt.title("Mon graphique")
plt.xlabel("Action")
plt.ylabel("Date")
plt.legend()

#plt.show()
```

```python
#histogramme
#représenter sur le même histogramme deux distributions issues de nombres aléatoires
plt.hist(np.random.randn(10000),bins=40,color="red",label="dist1",alpha=0.5)
plt.hist(np.random.randn(10000),bins=40,color="blue",label="dist2",alpha=0.5)
plt.legend()
```
